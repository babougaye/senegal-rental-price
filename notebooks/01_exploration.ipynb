# Exploration des données — Locations Senegal

Analyse exploratoire du jeu de données synthétique. Ce notebook ne contient **aucune logique de production** : il sert uniquement à comprendre les données avant d'écrire le pipeline dans `src/`.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/raw/locations_senegal.csv")
df.head()
```

```python
df.info()
df.isna().mean().sort_values(ascending=False)
```

```python
df["prix_loyer_mensuel"].describe()
```

```python
df.groupby("ville")["prix_loyer_mensuel"].median().sort_values(ascending=False).plot(kind="bar", title="Prix médian par ville")
plt.ylabel("FCFA / mois")
plt.show()
```

```python
dakar = df[df["ville"] == "Dakar"]
dakar.groupby("quartier")["prix_loyer_mensuel"].median().sort_values(ascending=False).plot(kind="bar", title="Prix médian par quartier (Dakar)")
plt.ylabel("FCFA / mois")
plt.show()
```

```python
plt.scatter(df["surface_m2"], df["prix_loyer_mensuel"], alpha=0.3)
plt.xlabel("Surface (m2)")
plt.ylabel("Prix loyer mensuel (FCFA)")
plt.title("Prix vs surface")
plt.show()
```

## Conclusions

- Le prix est fortement corrélé à la surface et à la ville (Dakar nettement plus cher).
- Les quartiers premium de Dakar (Almadies, Ngor, Plateau) confirment la prime attendue.
